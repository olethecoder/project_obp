```python
import unittest
import numpy as np
import pandas as pd
from input_parser import InputParser

class UnitTest():

    def __init__(self, shifts_df, tasks_df):
        super().__init__() 
        self.shifts = shifts_df[shifts_df["usage"] != 0]
        self.tasks = tasks_df
        self.days = ["monday","tuesday","wednesday","thursday","friday","saturday","sunday"]
        self.shifts_coverage = np.zeros((7, 24 * 4), dtype=int)
        self.tasks_coverage = np.zeros((7, 24 * 4), dtype=int)
        self.n  = self.shifts_coverage.shape[1]
        

    @staticmethod
    def times_to_quarter_index(time_input):
        """Convert a single time string or Pandas Series of time strings to a quarter-hour index."""
        
        def convert(time_str):
            """Helper function to convert a single time string."""
            if ':' in time_str:
                hours, minutes = map(int, time_str.split(':'))
            else:
                hours, minutes = map(int, time_str.split('.'))
            
            return (hours * 4) + (minutes // 15)
        
        # Check if the input is a Series
        if isinstance(time_input, pd.Series):
            return time_input.map(convert)  # Apply the function to each element
        else:
            return convert(time_input)  # Process single scalar value

    @staticmethod
    def quarter_index_to_time(index):
        """Convert a single quarter-hour index to a time string."""
        hours = index // 4
        minutes = (index % 4) * 15
        return f"{hours:02d}:{minutes:02d}"
    
    #gets length in time blocks of quarters. e.g. 30 --> 2, 45 --> 3
    @staticmethod
    def time_length(length):
        return int(length // 15)

       
    def same_end_day(self):
        self.shifts["same_end_day"] = self.times_to_quarter_index(self.shifts["end"]) > self.times_to_quarter_index(self.shifts["start"])
        self.tasks["same_end_day"] = self.times_to_quarter_index(self.tasks["end_window"]) > self.times_to_quarter_index(self.tasks["start_window"])
                

    #It is assumed that shifts are always shorter than 24 hours
    def shifts_coverage_matrix(self):
        """
        For each row in self.shifts, fill the coverage matrix so that 
        each quarter-hour that the shift is worked is incremented by its usage.
        The break quarter-hours are excluded from coverage.
        """
        # Clear coverage matrix first
        self.shifts_coverage = np.zeros((7, 24 * 4), dtype=int)

        for _, shift in self.shifts.iterrows():
            usage = int(shift["usage"])
            # If usage == 0, skip entirely
            if usage == 0:
                continue

            # For each day, if shift[day] == 1, we add coverage
            for day in self.days:
                if shift[day] == 1:
                    day_idx = self.days.index(day)
                    next_day_idx = (day_idx + 1) % 7  # wrap around after sunday to monday

                    # Quarter-hour indices
                    start_q = self.times_to_quarter_index(shift["start"])
                    end_q   = self.times_to_quarter_index(shift["end"])
                    brk_q   = self.times_to_quarter_index(shift["break"])
                    brk_len = self.time_length(shift["break_duration"])  # in quarter-hours
                    brk_end_q = brk_q + brk_len

                    # Because shift < 24h, check if it's "same-day" or "overnight"
                    if start_q < end_q:
                        # === Same Day Shift ===
                        # We'll build up all covered quarters ignoring the break.
                        covered_quarters = list(range(start_q, end_q))
                        # Remove the break slice [brk_q, brk_end_q) if it fits in [start_q, end_q)
                        # Make sure we don't remove beyond the shift's range
                        break_low  = max(start_q, brk_q)
                        break_high = min(end_q, brk_end_q)

                        # Filter out break quarters
                        covered_quarters = [
                            q for q in covered_quarters 
                            if not (break_low <= q < break_high)
                        ]
                        
                        # Now add coverage for those quarters
                        for q in covered_quarters:
                            self.shifts_coverage[day_idx, q] += usage

                    else:
                        # === Overnight Shift ===
                        # e.g. start_q=88 (22:00), end_q=24 (06:00)

                        # 1) coverage from start_q to midnight (index=96) in day_idx
                        # 2) coverage from 0 to end_q in next_day_idx
                        # We'll skip the break portion depending on where it lies.

                        # Part A: range day_idx from start_q..96
                        day_quarters = list(range(start_q, 96))
                        # Part B: range next_day_idx from 0..end_q
                        next_day_quarters = list(range(0, end_q))

                        # We'll remove break quarters from these slices
                        # Because 'break' could be interpreted as the same local day 
                        # or the 'next' day if brk_q < start_q means the break is 
                        # actually after midnight. So let's define a small helper:
                        
                        def remove_break(q_list, day_type="same"):
                            """
                            day_type = "same" means break is also in the same day coords
                            day_type = "next" means break is in the 'next day' coords
                            """
                            # If break is on the 'same' day: skip [brk_q, brk_end_q)
                            # If break is on the 'next' day: skip [brk_q, brk_end_q) from next_day coords
                            return [
                                q for q in q_list 
                                if not (brk_q <= q < brk_end_q)
                            ]

                        # We must decide if break is in day_idx coords or next_day_idx coords
                        # Normally, if brk_q >= start_q, that implies the break is still 
                        # on the same "day" portion. 
                        # If brk_q < start_q, that implies the break is actually 
                        # on the next day portion (since the clock "resets" at midnight).
                        
                        # handle break for 'day_quarters'
                        if brk_q >= start_q and brk_q < 96:
                            # break is in the first day portion
                            day_quarters = remove_break(day_quarters, day_type="same")

                        # handle break for 'next_day_quarters'
                        # If break indices are < start_q, we interpret that as break in the next day portion
                        # Alternatively, if brk_q >= 96, we definitely interpret it as next day. 
                        # The easiest robust approach is:
                        # shift the break indices by -96 if you want to treat next-day coverage from 0..end_q
                        # but let's do something simpler:
                        
                        # If brk_q < start_q => break is definitely on the next day.
                        # Then for the next-day coverage, we should treat the break as 
                        # [brk_q, brk_end_q) mapped into 0..(some number).
                        # So we define:
                        brk_q_next    = (brk_q - 96) if (brk_q >= 96) else brk_q
                        brk_end_q_next= (brk_end_q - 96) if (brk_end_q >= 96) else brk_end_q
                        
                        # Now anything in [0..end_q) that also lies in [brk_q_next, brk_end_q_next) 
                        # is break time in the next day.
                        next_day_quarters = [
                            q for q in next_day_quarters
                            if not (brk_q_next <= q < brk_end_q_next)
                        ]

                        # Add coverage
                        for q in day_quarters:
                            self.shifts_coverage[day_idx, q] += usage
                        for q in next_day_quarters:
                            self.shifts_coverage[next_day_idx, q] += usage


    def tasks_coverage_matrix(self):
        for _, task in self.tasks.iterrows():
            if task["duration"] == 0:
                continue 
            
            day_idx = task["day_index"]  # Now using the direct day_index from the task
            start_tasks_idx = int(self.times_to_quarter_index(task["solution_start"]))
            end_tasks_idx = int(start_tasks_idx + self.time_length(task["duration"]))
            required_nurses = int(task["required_nurses"])
            
            if end_tasks_idx < self.n:
                self.tasks_coverage[day_idx, start_tasks_idx : end_tasks_idx] += required_nurses
            else:
                overlap = end_tasks_idx - self.n
                self.tasks_coverage[day_idx, start_tasks_idx : self.n] += required_nurses
                next_day_idx = 0 if day_idx == 6 else day_idx + 1  # Using 6 for Sunday since it's zero-based
                self.tasks_coverage[next_day_idx, 0 : overlap] += required_nurses


    def check_coverage_sufficiency(self):
        
        shortages = self.tasks_coverage > self.shifts_coverage  # Boolean mask

        if np.any(shortages):
            print("Coverage shortages detected at these times and days:")
            
            # Iterate through shortages to display where the issue occurs
            for day_idx in range(7):
                for time_idx in range(24 * 4):
                    if shortages[day_idx, time_idx]:  # If shortage exists
                        day_name = self.days[day_idx]
                        time_str = self.quarter_index_to_time(time_idx)
                        print(f" - {day_name} at {time_str}: "
                            f"Needed {self.tasks_coverage[day_idx, time_idx]}, "
                            f"Available {self.shifts_coverage[day_idx, time_idx]}")
            
            return False  # There are coverage shortages
        else:
            print("All shifts fully cover the task requirements.")
            return True  # No shortages
        
    def check_solution_times(self):
    
        def is_time_in_window(row):
            start = row['start_window']
            end = row['end_window']
            solution = row['solution_start']
            same_day = row['same_end_day']
            
            if same_day:
                return (solution >= start) and (solution <= end)
            else:
                if solution >= start or solution <= end:
                    return True
                return False
        
        # Add validation column to existing DataFrame
        self.tasks['is_valid'] = self.tasks.apply(is_time_in_window, axis=1)
        
        # Check if any tasks are invalid
        invalid_tasks = self.tasks[~self.tasks['is_valid']]
        all_valid = len(invalid_tasks) == 0
        
        if all_valid:
            message = "All tasks start within their designated time windows."
            return True, message
        else:
            # Create detailed error message for invalid tasks
            invalid_details = []
            for _, task in invalid_tasks.iterrows():
                if task['same_end_day']:
                    window = f"[{task['start_window']}-{task['end_window']}]"
                else:
                    window = f"[{task['start_window']}-96] or [0-{task['end_window']}]"
                invalid_details.append(f"Task started at {task['solution_start']} but window is {window}")
            
            message = f"Found {len(invalid_tasks)} task(s) with invalid start times:\n" + "\n".join(invalid_details)
            return False, message
    

    #def all_tasks_covered(self):
        
unit_test = UnitTest(shifts_ole, tasks_ole)  # Ensure shifts and tasks coverage matrices are populated
unit_test.shifts_coverage_matrix()
unit_test.tasks_coverage_matrix()
unit_test.check_coverage_sufficiency()
unit_test.same_end_day()
unit_test.check_solution_times()
```

```python
import numpy as np
import pandas as pd

class UnitTest():
    def __init__(self, shifts_df, tasks_df):
        super().__init__() 
        print("\nInitializing Unit Test...")

        self.shifts = shifts_df[shifts_df["usage"] != 0].copy()
        self.tasks = tasks_df
        self.days = ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]
        self.shifts_coverage = np.zeros((7, 24 * 4), dtype=int)
        self.tasks_coverage = np.zeros((7, 24 * 4), dtype=int)
        self.n = self.shifts_coverage.shape[1]
        
        print(f"Loaded {len(self.shifts)} active shifts and {len(self.tasks)} tasks")

    @staticmethod
    def times_to_quarter_index(time_input):
        def convert(time_str):
            try:
                if ':' in time_str:
                    hours, minutes = map(int, time_str.split(':'))
                else:
                    hours, minutes = map(int, time_str.split('.'))
                
                if not (0 <= hours < 24 and 0 <= minutes < 60):
                    raise ValueError(f"Invalid time: {time_str}")
                    
                return (hours * 4) + (minutes // 15)
            except Exception as e:
                print(f"Error converting time '{time_str}': {str(e)}")
                raise
        
        if isinstance(time_input, pd.Series):
            return time_input.map(convert)
        else:
            return convert(time_input)

    @staticmethod
    def quarter_index_to_time(index):
        hours = index // 4
        minutes = (index % 4) * 15
        return f"{hours:02d}:{minutes:02d}"
    
    @staticmethod
    def time_length(length):
        return int(length // 15)
       
    def same_end_day(self):
        print("\nCalculating same-day end flags...")
        try:
            self.shifts["same_end_day"] = self.times_to_quarter_index(self.shifts["end"]) > self.times_to_quarter_index(self.shifts["start"])
            self.tasks["same_end_day"] = self.times_to_quarter_index(self.tasks["end_window"]) > self.times_to_quarter_index(self.tasks["start_window"])
            print("Successfully calculated same-day flags")
        except Exception as e:
            print(f"Error in same_end_day calculation: {str(e)}")
            raise

    def shifts_coverage_matrix(self):
        print("\nCalculating shifts coverage matrix...")
        for _, shift in self.shifts.iterrows():
            for day in self.days:
                if shift[day] == 1:
                    day_idx = self.days.index(day)
                    try:
                        start_shifts_idx = self.times_to_quarter_index(shift["start"])
                        break_shifts_idx = self.times_to_quarter_index(shift["break"])
                        end_shifts_idx = self.times_to_quarter_index(shift["end"])
                        break_length = self.time_length(shift["break_duration"])
                        usage = int(shift["usage"])

                        if start_shifts_idx <= break_shifts_idx and start_shifts_idx < end_shifts_idx:
                            self.shifts_coverage[day_idx, start_shifts_idx : break_shifts_idx] += usage
                            self.shifts_coverage[day_idx, break_shifts_idx + break_length : end_shifts_idx] += usage

                        elif start_shifts_idx <= break_shifts_idx and break_shifts_idx + break_length < self.n and start_shifts_idx >= end_shifts_idx:
                            self.shifts_coverage[day_idx, start_shifts_idx : break_shifts_idx] += usage
                            self.shifts_coverage[day_idx, break_shifts_idx : self.n] += usage
                            next_day_idx = 0 if day == "sunday" else day_idx + 1
                            self.shifts_coverage[next_day_idx, 0 : end_shifts_idx] += usage

                        elif start_shifts_idx <= break_shifts_idx and break_shifts_idx + break_length >= self.n and start_shifts_idx >= end_shifts_idx:
                            overlap = break_shifts_idx + break_length - self.n
                            self.shifts_coverage[day_idx, start_shifts_idx : break_shifts_idx] += usage
                            next_day_idx = 0 if day == "sunday" else day_idx + 1
                            self.shifts_coverage[next_day_idx, overlap : end_shifts_idx] += usage

                        else:
                            self.shifts_coverage[day_idx, start_shifts_idx : self.n] += usage
                            next_day_idx = 0 if day == "sunday" else day_idx + 1
                            self.shifts_coverage[next_day_idx, 0 : break_shifts_idx] += usage
                            self.shifts_coverage[next_day_idx, break_shifts_idx + break_length : end_shifts_idx] += usage
                    
                    except Exception as e:
                        print(f"Error processing shift on {day}: {str(e)}")
                        print(f"Shift details: start={shift['start']}, break={shift['break']}, end={shift['end']}")
                        raise
        print(self.shifts_coverage)               
        print("Successfully calculated shifts coverage matrix")

    def tasks_coverage_matrix(self):
        print("\nCalculating tasks coverage matrix...")
        for idx, task in self.tasks.iterrows():
            if task["duration"] == 0:
                print(f"Skipping task {idx} with zero duration")
                continue 
            
            try:
                day_idx = int(task["day_index"])
                if not 0 <= day_idx <= 6:
                    raise ValueError(f"Invalid day index {day_idx} for task {idx}")
                
                start_tasks_idx = int(self.times_to_quarter_index(task["solution_start"]))
                end_tasks_idx = int(start_tasks_idx + self.time_length(task["duration"]))
                
                if end_tasks_idx < self.n:
                    self.tasks_coverage[day_idx, start_tasks_idx : end_tasks_idx] += task["required_nurses"]
                else:
                    overlap = end_tasks_idx - self.n
                    self.tasks_coverage[day_idx, start_tasks_idx : self.n] += task["required_nurses"]
                    next_day_idx = 0 if day_idx == 6 else day_idx + 1
                    self.tasks_coverage[next_day_idx, 0 : overlap] += task["required_nurses"]
            
            except Exception as e:
                print(f"Error processing task {idx}: {str(e)}")
                print(f"Task details: day={task['day_index']}, start={task['solution_start']}, duration={task['duration']}")
                raise
        print(self.tasks_coverage)
        print("Successfully calculated tasks coverage matrix")

    def check_coverage_sufficiency(self):
        print("\nChecking coverage sufficiency...")
        shortages = self.tasks_coverage > self.shifts_coverage

        if np.any(shortages):
            print("\nCoverage shortages detected at these times and days:")
            shortage_count = 0
            for day_idx in range(7):
                day_shortages = []
                for time_idx in range(24 * 4):
                    if shortages[day_idx, time_idx]:
                        shortage_count += 1
                        day_shortages.append(
                            f"    {self.quarter_index_to_time(time_idx)}: "
                            f"Need {self.tasks_coverage[day_idx, time_idx]} nurses, "
                            f"Have {self.shifts_coverage[day_idx, time_idx]}"
                        )
                if day_shortages:
                    print(f"\n  {self.days[day_idx].capitalize()}:")
                    print("\n".join(day_shortages))
            
            print(f"\nTotal shortage points: {shortage_count}")
            return False
        else:
            print("All shifts fully cover the task requirements ")
            return True

    def check_solution_times(self):
        print("\nChecking if all tasks start within their time windows...")
    
        def is_time_in_window(row):
            # Handle both string and integer inputs
            def get_index(value):
                if isinstance(value, str):
                    return self.times_to_quarter_index(value)
                return int(value)  # Already an index
            
            start = get_index(row['start_window'])
            end = get_index(row['end_window'])
            solution = get_index(row['solution_start'])
            same_day = row['same_end_day']
            
            if same_day:
                return (solution >= start) and (solution <= end)
            else:
                if solution >= start or solution <= end:
                    return True
                return False

        try:
            self.tasks['is_valid'] = self.tasks.apply(is_time_in_window, axis=1)
            
            invalid_tasks = self.tasks[~self.tasks['is_valid']]
            all_valid = len(invalid_tasks) == 0
            
            if all_valid:
                print("All tasks start within their designated time windows ")
                return True, "All tasks start within their designated time windows."
            else:
                print(f"\nFound {len(invalid_tasks)} task(s) with invalid start times:")
                invalid_details = []
                for idx, task in invalid_tasks.iterrows():
                    if task['same_end_day']:
                        window = f"[{task['start_window']}-{task['end_window']}]"
                    else:
                        window = f"[{task['start_window']}-96] or [0-{task['end_window']}]"
                    invalid_details.append(f"Task {idx}: Starts at {task['solution_start']} but window is {window}")
                
                message = "\n".join(invalid_details)
                print(message)
                return False, message
                
        except Exception as e:
            print(f"Error checking solution times: {str(e)}")
            raise

unit_test = UnitTest(shifts_ole, tasks_ole)  # Ensure shifts and tasks coverage matrices are populated
unit_test.shifts_coverage_matrix()
unit_test.tasks_coverage_matrix()
unit_test.check_coverage_sufficiency()
unit_test.same_end_day()
unit_test.check_solution_times()
```
